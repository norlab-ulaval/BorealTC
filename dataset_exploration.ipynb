```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
data_dir = Path("data")
terrains = [f.stem for f in data_dir.iterdir() if f.is_dir()]
```

```python
fig, subfigs = plt.subplots(nrows=2, ncols=2, sharey=True)

for ax, terr in zip(subfigs.flatten(), terrains):
    imu_df = pd.read_csv(data_dir / terr / "imu_1.csv")
    pro_df = pd.read_csv(data_dir / terr / "pro_1.csv")

    pro_df.Lcur.plot(ax=ax, label="Left")
    pro_df.Rcur.plot(ax=ax, label="Right")
    ax.set_title(terr)
    ax.set_ylabel("Current (A)")

fig.tight_layout()
```
